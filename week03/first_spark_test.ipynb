```python
from pyspark.sql import functions as F

# Use Unity Catalog volume path
df = (
    spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .csv("/Volumes/<catalog>/<schema>/<volume>/hospitals.csv")
)

result = (
    df.groupBy("city")
      .agg(
        F.count("*").alias("num_hospitals"),
        F.sum("beds").alias("total_beds")
      )
)

result.show()
```
